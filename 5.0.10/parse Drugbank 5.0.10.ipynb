```python
import csv
import xml.etree.ElementTree as ET
import time
import io
from tqdm import tqdm
import pandas as pd
tree = ET.parse("C:/Drugbank/5.0.11/full database.xml")
```

```python
root = tree.getroot()
```

```python
group_focus = "approved"
drugs = list(root)
with io.open('C:/Drugbank/5.0.11/ddi-5.0.10.csv', "w", encoding="utf-8", newline='') as csv_file3:
    results_writer_cd = csv.writer(csv_file3, delimiter=',', quotechar='"', quoting=csv.QUOTE_MINIMAL)
    for i in tqdm(range(len(drugs))):
    #for i in range(0, 10):
        drug = drugs[i]
        idDB = drug[0].text
        for idx,feature in enumerate(drug):
            if 'name' in str(feature): # drug name
                drug_name = drug[idx].text
                #print(drug_name)
            if 'groups' in str(feature): #type of drug
                group_name = list(drug[idx])[0].text
                #print(group_name)
            if 'drug-interactions' in str(feature): #interaction other drugs
                #drug_interaction_ids = ';'.join([di[0].text for di in list(drug[idx])])
                drug_interaction = ';'.join([di[0].text for di in list(drug[idx])])
                #print(drug_interaction)
        if (group_focus in group_name):
            #print (" step ", i)
            print(idDB, drug_name, group_name, drug_interaction)
            results_writer_cd.writerow([idDB, drug_name, group_name, drug_interaction])
```

```python
with open("C:/Drugbank/5.0.11/ddi-5.0.11-interactions.csv", 'r') as f:
    reader = csv.reader(f)
    my_list = list(reader)
```

```python
interactions_raw = []
for iter in range(len(my_list)):
    if my_list[iter][1] != '':
        listinter = my_list[iter][1].split (";")
        for it in range (len(listinter)):
            interactiunea = [my_list[iter][0], listinter[it]]
            interactiunea.sort()
            interactions_raw.append(interactiunea)
print (interactions_raw)
```

```python
interactions_list = []
for elem in interactions_raw:
    if elem not in interactions_list:
        interactions_list.append(elem)
```

```python
print (interactions_list)
```

```python
with io.open('C:/Drugbank/5.0.11/ddi-5.0.11-edges.csv', "w", encoding="utf-8", newline='') as csv_file4:
    results_writer_edge = csv.writer(csv_file4, delimiter=',', quotechar='"', quoting=csv.QUOTE_MINIMAL)
    for i in tqdm(range(len(interactions_list))):
        results_writer_edge.writerow([interactions_list[i][0], interactions_list[i][1]])
```
