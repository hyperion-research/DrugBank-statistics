```python
import csv
import time
import io
import pandas as pd
from tqdm import tqdm
import pandas as pd
with open("C:/Drugbank/5.1.8/ddi-5.1.8-atc-id-name.csv", 'r') as f:
    reader = csv.reader(f)
    my_list = list(reader)
with open("C:/DrugAge/n8.csv", 'r') as fi:
    readeri = csv.reader(fi)
    my_drugs_age = list(readeri)
```

```python
with io.open('C:/DrugAge/name-DB-2.csv', "w", encoding="utf-8", newline='') as csv_file4:
    results_writer_edge = csv.writer(csv_file4, delimiter=',', quotechar='"', quoting=csv.QUOTE_MINIMAL)
    for i in range(0, len(my_drugs_age)):
        dn = my_drugs_age[i][0]
        dc = "not found"
        for j in range(0, len(my_list)):
            if (dn == my_list[j][1]):
                dc = my_list[j][0]
        results_writer_edge.writerow([dn, dc])
```
