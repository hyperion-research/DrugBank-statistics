```python
import csv
import xml.etree.ElementTree as ET
import time
import io
from tqdm import tqdm
import pandas as pd
tree = ET.parse("C:/Drugbank/5.0.11/full database.xml")
```

```python
root = tree.getroot()
```

```python
group_focus = "approved"
drugs = list(root)
with io.open('C:/Drugbank/5.0.11/dti-5.0.11.csv', "w", encoding="utf-8", newline='') as csv_file3:
    results_writer_cd = csv.writer(csv_file3, delimiter=',', quotechar='"', quoting=csv.QUOTE_MINIMAL)
    for i in tqdm(range(len(drugs))):
    #for i in range(0, 10):
        drug = drugs[i]
        idDB = drug[0].text
        for idx,feature in enumerate(drug):
            if 'name' in str(feature): # drug name
                drug_name = drug[idx].text
                #print(drug_name)
            if 'groups' in str(feature): #type of drug
                group_name = list(drug[idx])[0].text
                #print(group_name)
            if 'targets' in str(feature): #interaction other drugs
                #drug_interaction_ids = ';'.join([di[0].text for di in list(drug[idx])])
                drug_target_interaction = ';'.join([di[0].text for di in list(drug[idx])])
                #print(drug_interaction)
        if (group_focus in group_name):
            #print (" step ", i)
            print(idDB, drug_name, group_name, drug_target_interaction)
            results_writer_cd.writerow([idDB, drug_name, drug_target_interaction])
```

```python
with open("C:/Drugbank/5.0.11/dti-5.0.11-interactions.csv", 'r') as f:
    reader = csv.reader(f)
    my_list = list(reader)
```

```python
interactions_raw = []
for iter in range(len(my_list)):
    if my_list[iter][1] != '':
        listinter = my_list[iter][1].split (";")
        for it in range (len(listinter)):
            interactiunea = [my_list[iter][0], listinter[it]]
            #interactiunea.sort()
            interactions_raw.append(interactiunea)
print (interactions_raw)
```

```python
with io.open('C:/Drugbank/5.0.11/dti-5.0.11-edges.csv', "w", encoding="utf-8", newline='') as csv_file4:
    results_writer_edge = csv.writer(csv_file4, delimiter=',', quotechar='"', quoting=csv.QUOTE_MINIMAL)
    for i in tqdm(range(len(interactions_raw))):
        results_writer_edge.writerow([interactions_raw[i][0], interactions_raw[i][1]])
```

```python
interactions_bi = []
my_drugs = []
my_targets = []
for i in range (0, len(interactions_raw)):
    interactions_bi.append(tuple(interactions_raw[i]))
    my_drugs.append(interactions_raw[i][0])
    my_targets.append(interactions_raw[i][1])
#print (interactions_bi)
```

```python
import networkx as nx 
from networkx.algorithms import bipartite
BG = nx.Graph()
BG.add_nodes_from(my_drugs, bipartite=0)
BG.add_nodes_from(my_targets, bipartite=1)
BG.add_edges_from(interactions_bi)
```

```python
print (BG.edges())
```

```python
from networkx.algorithms import bipartite
CBG = bipartite.color(BG)
```

```python
nx.set_node_attributes(BG, CBG, "bipartite")
print(CBG)
```

```python
drugs_c = []
targets_c =[]
for key, value in CBG.items():
    if value == 0:
        drugs_c.append(key)
    else:
        targets_c.append(key)
print ("no. of drugs: ", len(drugs_c))
print ("no. of targets: ", len(targets_c))
print ("no. of links: ", len(interactions_raw))
```

```python
print (len(BG.edges()))
```

```python
print("No of nodes ", BG.number_of_nodes())
print("No of links ", BG.number_of_edges())
print("Density ", nx.density(BG))
giant = max(nx.connected_components(BG), key=len)
gcc = BG.subgraph(giant)
print("No Nodes largest component ", gcc.number_of_nodes())
print("Diameter of largest component ", nx.diameter(gcc))
degrees = BG.degree()
degree_values = [v for k, v in degrees]
sum_of_degrees = sum(degree_values)
average_degree = sum_of_degrees/BG.number_of_nodes()
print("Average degree ", average_degree)
print("Average path length ", nx.average_shortest_path_length(gcc))
print("Average clustering coefficient ", nx.average_clustering(BG))
```

```python
lnodesgc = list(giant)
no_drugs_gc = 0
no_targets_gc = 0
for i in range (0, len(lnodesgc)):
    if (lnodesgc[i][0] == 'D'):
        no_drugs_gc = no_drugs_gc + 1
    else:
        no_targets_gc = no_targets_gc + 1
print ("Drugs in GC: ", no_drugs_gc)
print ("Targets in GC: ", no_targets_gc)
```

```python
no_d = 0
no_t = 0
deg_d = 0
deg_t = 0
list_nodes = list(BG.nodes())
for i in range (0, len(list_nodes)):
    if (list_nodes[i][0] == 'D'):
        no_d = no_d + 1
        deg_d = deg_d + BG.degree(list_nodes[i])
    else:
        no_t = no_t + 1
        deg_t = deg_t + BG.degree(list_nodes[i])
print ("Drug average degree: ", deg_d/no_d)
print ("Target average degree: ", deg_t/no_t)
```

```python
import collections
import matplotlib.pyplot as plt 
degrees_all = [BG.degree(n) for n in BG.nodes()]
dict_dd = { i : degrees_all[i] for i in range(0, len(degrees_all) ) }
dd_2 = dict(collections.Counter(dict_dd.values()))
plt.xscale('log')
plt.yscale('log')
plt.title('Degree distribution')
plt.xlabel('Frequency')
plt.ylabel('Degree')
#plt.legend(['Initial network','Network after attack-repair rounds'], loc='upper right')
#plt.xlim((2.5e-3,1.5e-1))
#plt.ylim((.9,6.5e2))
plt.scatter(dd_2.keys(),dd_2.values(),c='b',marker='x')
plt.legend(['DTI network'], loc='upper right')
```

```python
list_nodes_2 = list(BG.nodes())
degrees_all_d = []
degrees_all_t = []
for i in range (0, len(list_nodes_2)):
    if (list_nodes_2[i][0] == 'D'):
        degrees_all_d.append(BG.degree(list_nodes_2[i]))
    else:
        degrees_all_t.append(BG.degree(list_nodes_2[i]))

dict_ddd = { i : degrees_all_d[i] for i in range(0, len(degrees_all_d) ) }
dd_2d = dict(collections.Counter(dict_ddd.values()))
plt.xscale('log')
plt.yscale('log')
plt.title('Drug degree distribution')
plt.xlabel('Frequency')
plt.ylabel('Degree')
#plt.legend(['Initial network','Network after attack-repair rounds'], loc='upper right')
#plt.xlim((2.5e-3,1.5e-1))
#plt.ylim((.9,6.5e2))
plt.scatter(dd_2d.keys(),dd_2d.values(),c='b',marker='x')
plt.legend(['DTI network'], loc='upper right')
```

```python
import numpy as np
import powerlaw
deg_array_all = np.asarray(degrees_all)
results_deg_all = powerlaw.Fit(deg_array_all)
print ("Degree all")
print(results_deg_all.power_law.alpha)
print(results_deg_all.power_law.xmin)
deg_array_all_d = np.asarray(degrees_all_d)
results_deg_all_d = powerlaw.Fit(deg_array_all_d)
print ("Degree drugs")
print(results_deg_all_d.power_law.alpha)
print(results_deg_all_d.power_law.xmin)
deg_array_all_t = np.asarray(degrees_all_t)
results_deg_all_t = powerlaw.Fit(deg_array_all_t)
print ("Degree targets")
print(results_deg_all_t.power_law.alpha)
print(results_deg_all_t.power_law.xmin)
```
